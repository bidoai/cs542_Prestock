Model reads the webscrapped data from the CSV and trains our Keras LSTM Model

```python
# importing data 
df1 = pd.read_csv('data_training.csv', index_col  = False)
```

```python

X=((df1["X"]))


max_1 = 0
for i in X:
    if len(i)>=max_1:
        max_1 = len(i)

X = [ast.literal_eval(x) for x in X]
X_temp = X  
```

```python
#padding
 
X = keras.preprocessing.sequence.pad_sequences(X, maxlen=max_1,dtype="float32")
```

```python
y_labels = pd.read_csv('y_labels.csv',header=None)
y_labels = y_labels.drop(0, 1)
```

```python
y_labels = np.array(y_labels)
```

```python
#y_labels =y_labels.tolist()
```

```python
y_labels
```

```python
#reshaping inputs
X = X.reshape(X.shape[0],1,X.shape[1])
y_labels = y_labels.reshape(y_labels.shape[0],1)
```

```python
#our stacked LSTM model
X_train, X_val, y_train, y_val = train_test_split(X, 
                                                    y_labels, 
                                                    test_size=0.1, 
                                                    random_state=1)


data_dim= max_1


model = Sequential()
model.add(LSTM(32, return_sequences=True,
               input_shape=(1, data_dim)))  # returns a sequence of vectors of dimension 32

model.add(LSTM(32, return_sequences=True))  # returns a sequence of vectors of dimension 32
model.add(LSTM(32))  # return a single vector of dimension 32
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])




model.fit(X_train, y_train,
          batch_size=32, epochs=300,
          validation_data=(X_val, y_val))

my_predict = model.predict_classes(X)
```

```python
model.summary()
```

```python

#saving the model to predict
model.save("LSTM_Model.h5py")
```

```python
score = model.evaluate(X, y_labels, verbose=0)
print("%s: %.2f%%" % (model.metrics_names[1], score[1]*100))
```
